```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('../Corona_NLP_train.csv', encoding='latin1')
```

```python
dataset.head()
```

```python
import re

texts=[]
for i in range(0, len(dataset)):
    text = re.sub('[^a-zA-Z]', ' ', dataset['OriginalTweet'][i])
    text = text.lower()
    text = text.split()
    x = len(text) if text.count('https') == 0 else text.index('https')
    text = [t for t in text if not t=='https']
    text = ' '.join(text)
    texts.append(text)
    
```

```python
print(len(texts))
```

```python
texts
```

# Training the word2vec model

```python
from gensim.models import Word2Vec
```

```python
sentences = [line.split() for line in texts]

w2v = Word2Vec(sentences, vector_size=100, window=5, workers=2, epochs=10, min_count=5) 
```

```python
print(sentences[20:25])
```

```python
words = list(w2v.wv.key_to_index)
```

```python
print(words)
```

```python
print(len(words))
```
